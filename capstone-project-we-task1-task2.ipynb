```python
import pixellib
from pixellib.instance import instance_segmentation
import cv2
```

```python
segment_video = instance_segmentation()
segment_video.load_model("mask_rcnn_coco.h5")
VIDEO_PATH = "/kaggle/input/video-final-project/TahrirSquare.mp4"
```

```python
segment_video.process_video(VIDEO_PATH, show_bboxes = True, frames_per_second= 20, output_video_name="Tahrir_Square_Detected_All.mp4")
```

```python
target_classes = segment_video.select_target_classes(car=True)
```

```python
segment_video.process_video(VIDEO_PATH, segment_target_classes= target_classes,show_bboxes = True, frames_per_second= 20, output_video_name="Tahrir_Square_Cars_Only.mp4")
```

```python
cap = cv2.VideoCapture(VIDEO_PATH)

# Get the video's frames per second (fps) and frame dimensions
fps = cap.get(cv2.CAP_PROP_FPS)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Define the codec and create VideoWriter object
output_file = "Car_Count_Frame_by_Frame.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_file, fourcc, 20, (frame_width, frame_height))

# Initialize the car counter
car_count = 0
i = 1

# Loop over the frames in the video
while cap.isOpened():
    # Get the next frame
    ret, frame = cap.read()
    if frame is None:
        break
    
    # Segment the frame
    segmentation_result = segment_video.segmentFrame(frame,segment_target_classes= target_classes,show_bboxes = True)
    class_ids = segmentation_result[0]['class_ids'].astype(int)
    
    # Count the number of cars in the frame
    for class_id in class_ids:
        if class_id == 3:
            car_count += 1
    
    # Put the car count on the frame
    cv2.putText(frame, f"Car Count: {car_count}", (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    # Write the frame to the output video
    out.write(frame)
    
    print(f"No of frames:{i}")
    i += 1
    car_count = 0
    
    # Press `q` to quit
    key = cv2.waitKey(1)
    if key == ord("q"):
        break

# Release VideoWriter and close the video file
out.release()
cap.release()
```

```python
from scipy.spatial import distance

# Open the video file
cap = cv2.VideoCapture(VIDEO_PATH)

# Define the output video writer
output_path = "Tracked_Cars.mp4"
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'XVID'), 20, (frame_width, frame_height))

# Initialize car coordinates and IDs
car_coords = []
car_ids = []
car_counter = 0
i = 1
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    # Perform instance segmentation on the current frame
    result = segment_video.segmentFrame(frame,segment_target_classes= target_classes,show_bboxes = True)
    output_frame = result[1]
    
    # Get the detected car bounding boxes and IDs in the current frame
    car_bboxes = result[0]['rois']
    car_class_ids = result[0]['class_ids']
    
    # Update the list of car coordinates and IDs
    new_car_coords = []
    new_car_ids = []
    for bbox, class_id in zip(car_bboxes, car_class_ids):
        y1, x1, y2, x2 = bbox
        center = ((x1 + x2) / 2, (y1 + y2) / 2)
        new_car_coords.append(center)
        new_car_ids.append(class_id)
    
    # Compare with previous frame and count new cars
    new_car_counter = 0
    for coord, car_id in zip(new_car_coords, new_car_ids):
        if all(distance.euclidean(coord, prev_coord) > 55 for prev_coord in car_coords):
            car_counter += 1
            new_car_counter += 1
            car_coords.append(coord)
            car_ids.append(car_id)
    
    # Display the car counter on the frame
    cv2.putText(output_frame, f"Tracked Cars: {car_counter}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    # Write the processed frame to the output video
    out.write(output_frame)
    print(f"No. of Frames {i}")
    i+=1
    # Exit when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and writer objects
cap.release()
out.release()
cv2.destroyAllWindows()
```
